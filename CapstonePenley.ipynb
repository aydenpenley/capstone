```python
total_cycles = train.groupby('unit')['cycles'].max()
```

```python
train = train.merge(total_cycles, on='unit', suffixes=('', '_total'))
train = train.loc[:, ~train.columns.duplicated()]

print(train[['unit', 'cycles_total']].head())

train['RUL'] = train['cycles_total'] - train['cycles']
train.head()
```

```python
total_cycles = test.groupby('unit')['cycles'].max()
```

```python
test = test.merge(total_cycles, on='unit', suffixes=('', '_total'))
test = test.loc[:, ~test.columns.duplicated()]

print(test[['unit', 'cycles_total']].head())

test['RUL'] = test['cycles_total'] - test['cycles']
test.head()
```

```python
X_train = train.drop(columns=['RUL', 'unit', 'cycles'])  
y_train = train['RUL']

X_test = test.drop(columns=['RUL', 'unit', 'cycles'])
y_test = test['RUL']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
import xgboost as xgb

xgb_model = xgb.XGBRegressor(objective="reg:squarederror")

xgb_param_grid = {
    'learning_rate': [0.1, 0.01],
    'max_depth': [3, 5, 7],
    'n_estimators': [150, 200],
    'subsample': [0.8, 1.0]
}

grid_search = GridSearchCV(estimator=xgb_model, param_grid=xgb_param_grid, cv=5, scoring='neg_mean_absolute_error')
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```

* Given that the max amount of cycles recorded in the training set was 362, the MAE being about 44 isn't the greatest in this context and could be improved upon. 

```python
residuals = y_test - y_pred
plt.hist(residuals, bins=20)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```
